```python
df = data.drop(['race/ethnicity', 'parental level of education'], axis = 1)
df
```

```python
corl = df.corr()
corl
```

```python
plt.figure(figsize = (12, 8))
plt.scatter(df['math score'], df['reading score'])
plt.xlabel('math score')
plt.ylabel('reading score')
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
sns.countplot(df['math score'], palette = 'dark')
plt.title('Math Score',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(131)
plt.title('Math Scores')
sns.barplot(x="gender", y="math score", data=df)
```

```python
plt.figure()
plt.hist(df['math score'])
plt.xticks(range(10))
```

```python
plt.figure(figsize = (10, 10))
plt.gcf().text(.5, .9, 'Scatter Plot', fontsize = 40, color = 'Black', ha = 'center', va = 'center')
sns.scatterplot(x = 'reading score', y = 'writing score', hue = 'math score', data = df)
```

```python
plt.figure()
x = df['reading score']
y = df['writing score']
plt.bar(x, y, width = 0.4, color = 'yellow')
plt.bar(x + 0.5, y, width = 0.4, color = 'green')
```

```python
plt.figure()
x = df['reading score'].head(100)
y = df['writing score'].head(100)
plt.plot(x, 'o-', y, '-o')
plt.xlabel('reading score')
plt.ylabel('writing score')
plt.legend(['reading score', 'writing score'])
```
